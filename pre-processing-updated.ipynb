```python
import json
import pandas as pd
```

```python
with open('new-fine.json') as f:
  data = json.load(f)
```

```python
df = pd.DataFrame(data, columns=["Country", "Authority", "Date", "Fine", "Controller", "Category", "QuotedArt", "Type", "Summary", "Info"])
```

```python
df = df[df.Fine != 'Unknown']
df = df[df.Type != 'Unknown']
```

```python
df[['Fine']] = df[['Fine']].apply(lambda x: x.str.replace('Euro', '').replace(' ', ''))
df[['Fine']] = df[['Fine']].apply(lambda x: x.str.replace(' ', ''))
```

```python
df[['Fine']] = df[['Fine']].apply(lambda x: x.str.replace(',', ''))
df = df[df.Fine != 'ohnePunkte']

# df[['Fine']] = df[['Fine']].apply(lambda x: x.str.replace(',', '').astype(int))
```

```python
df[['Fine']] = df[['Fine']].apply(lambda x: x.str.replace('', '').astype(int))
# df.Fine.unique()
```

```python
df = df.reset_index(drop=True)
df.to_csv("new-fine-db.csv", index=False)
```
